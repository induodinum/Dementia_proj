```python
import pandas as pd
import numpy as np

from tqdm import tqdm
from os import listdir
from os.path import isfile, join
```

```python
mypath = '/Users/admin/Desktop/coding/senior_proj_resources/OpportunityUCIDataset/dataset'
files = [f for f in listdir(mypath) if isfile(join(mypath, f)) and f[-3:]=='dat' and f[-9:-4]!='Drill']
files = sorted(files)
files = [join(mypath,f) for f in files]

df1 = pd.read_csv(files[0], sep='\s+', usecols=[0,23,24,25,243],names=['ts','x','y','z','label'], header=None, engine='python')

xyz_cols = ['x','y','z']

tqdm.pandas(desc='Progress bar')
df1.progress_apply(lambda x: x**2)

df1 = df1.dropna(axis='index')
df1[xyz_cols] = df1[xyz_cols].apply(lambda x: x/1000 * 9.8)
```
